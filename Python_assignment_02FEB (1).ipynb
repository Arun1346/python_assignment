```python
#Q1. Explain with an example each when to use a for loop and a while loop.
''' For Loop:
Use a for loop when the number of iterations is known or you're iterating over a sequence (like a list, range, or string).

Example: '''
for num in range(1, 6):
    print(num)

''' While Loop:
Use a while loop when the number of iterations is unknown, and you want to loop until a condition becomes false.
Example: '''
num = 1
while num <= 5:
    print(num)
    num += 1
```

```python
#Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

''' using for loop  '''
# Initialize sum and product
sum_of_numbers = 0
product_of_numbers = 1

# Loop through the first 10 natural numbers (1 to 10)
for num in range(1, 11):
    sum_of_numbers += num
    product_of_numbers *= num

# Print the results
print(sum_of_numbers)
print(product_of_numbers)

```

```python
#Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

'''  using while loop  '''
# Initialize sum, product, and counter
sum_of_numbers = 0
product_of_numbers = 1
num = 1

# Loop until the counter reaches 10
while num <= 10:
    sum_of_numbers += num
    product_of_numbers *= num
    num += 1

#print the results
print(sum_of_numbers)
print(product_of_numbers)
```

```python
''' Q3.Create a python program to compute the electricity bill for a household.

The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
You are required to take the units of electricity consumed in a month from the user as input.


Your program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.      '''

def calculate_electricity_bill(units):
    # Initialize the total bill to 0
    bill = 0

    # First 100 units: ₹4.5 per unit
    if units <= 100:
        bill = units * 4.5
    # Next 100 units (101 to 200): ₹6 per unit
    elif units <= 200:
        bill = (100 * 4.5) + ((units - 100) * 6)
    # Next 100 units (201 to 300): ₹10 per unit
    elif units <= 300:
        bill = (100 * 4.5) + (100 * 6) + ((units - 200) * 10)
    # Above 300 units: ₹20 per unit
    else:
        bill = (100 * 4.5) + (100 * 6) + (100 * 10) + ((units - 300) * 20)

    return bill

# Example usage
units_consumed = int(input("Enter the number of units consumed: "))
total_bill = calculate_electricity_bill(units_consumed)
print(f"The total electricity bill for {units_consumed} units is: ₹{total_bill}")
```

```python
''' Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list. '''
#using for loop
# Create an empty list to store numbers whose cube is divisible by 4 or 5
result_list_for = []

# Iterate through numbers from 1 to 100
for num in range(1, 101):
    cube = num ** 3
    # Check if the cube is divisible by 4 or 5
    if cube % 4 == 0 or cube % 5 == 0:
        result_list_for.append(num)

# Print the resulting list
print("Numbers whose cube is divisible by 4 or 5 (for loop):", result_list_for)
```

```python
#using while loop
# Create an empty list to store numbers whose cube is divisible by 4 or 5
result_list_while = []

# Initialize a counter
num = 1

# Loop through numbers from 1 to 100
while num <= 100:
    cube = num ** 3
    # Check if the cube is divisible by 4 or 5
    if cube % 4 == 0 or cube % 5 == 0:
        result_list_while.append(num)
    num += 1

# Print the resulting list
print("Numbers whose cube is divisible by 4 or 5 (while loop):", result_list_while)
```

```python
''' Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"  '''

# Define the string
string = "I want to become a data scientist"

# Define vowels
vowels = "aeiouAEIOU"

# Initialize a counter for vowels
vowel_count = 0

# Loop through each character in the string and check if it's a vowel
for char in string:
    if char in vowels:
        vowel_count += 1

# Print the total number of vowels
print(f"The number of vowels in the string is: {vowel_count}")
```
